```python
import pandas as pd
from pathlib import Path
```

```python
# load the dataset in a Pandas DataFrame
file_path = Path("Resources/iris.csv")
iris_df = pd.read_csv(file_path)
iris_df.head()
```

```python
# Drop class column
new_iris_df= iris_df.drop(['class'], axis=1)
new_iris_df
```

```python
# Reordering the columns 
new_iris_df= new_iris_df[['sepal_length','petal_length','sepal_width','petal_width']]
new_iris_df
```

```python
# Export the new Dataframe to a csv
output_file_path = "Cruptocurrencies/Resources/new_iris_data.csv"
new_iris_df.to_csv(output_file_path, index=False)
```
